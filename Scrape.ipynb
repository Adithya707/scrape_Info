```python
#Requirements
#pip3 install requests
#pip3 install bs4

# import these two modules bs4 for selecting HTML tags easily
from bs4 import BeautifulSoup
# requests module is easy to operate some people use urllib but I prefer this one because it is easy to use.
import requests
```

```python
url="https://krishijagran.com/events/?loc=asia"
source=requests.get(url)

soup=BeautifulSoup(source.text,'html')

title=soup.find('title') # place your html tagg in parentheses that you want to find from html.
print("this is with html tags :",title)
print(title.text)
```

```python
#All the events listed for five months.
ww2_contents=soup.find_all("ul",class_='list-unstyled evt-list')

for i in ww2_contents:
    print(i.text)
```

```python
location = soup.findAll('ul',class_='list-unstyled evt-list')
for i in location: # getting each li tag where information is stored
    for m in i:
        print(m)
```
